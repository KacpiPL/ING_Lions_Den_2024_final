```python
def calculate_psi(expected, actual, buckets=10):
    """Calculate Population Stability Index (PSI) for two distributions."""
    
    def calculate_bucketed_counts(data, buckets):
        # Calculate bucket boundaries
        quantiles = np.linspace(0, 100, num=buckets + 1)
        bin_edges = np.percentile(data, q=quantiles)
        
        # Assign data points to buckets
        bucket_indices = np.digitize(data, bin_edges[1:-1])
        
        # Calculate counts for each bucket
        bucket_counts = pd.Series(bucket_indices).value_counts(sort=False)
        
        return bin_edges, bucket_counts
    
    def calculate_psi_bucket(expected_counts, actual_counts):
        expected_freq = expected_counts / expected_counts.sum()
        actual_freq = actual_counts / actual_counts.sum()
        
        # Calculate PSI for each bucket
        psi_bucket = (actual_freq - expected_freq) * np.log(actual_freq / expected_freq)
        
        return psi_bucket.sum()
    
    # Calculate PSI for each variable
    psi_values = {}
    for col in expected.columns:
        expected_data = expected[col].dropna()
        actual_data = actual[col].dropna()
        
        # Calculate bucketed counts for expected and actual data
        expected_bin_edges, expected_counts = calculate_bucketed_counts(expected_data, buckets)
        actual_bin_edges, actual_counts = calculate_bucketed_counts(actual_data, buckets)
        
        # Extend bin edges for consistent calculations
        all_bin_edges = np.unique(np.concatenate([expected_bin_edges, actual_bin_edges]))
        
        # Re-calculate bucketed counts with consistent bin edges
        expected_counts = pd.cut(expected_data, bins=all_bin_edges, include_lowest=True).value_counts(sort=False)
        actual_counts = pd.cut(actual_data, bins=all_bin_edges, include_lowest=True).value_counts(sort=False)
        
        # Calculate PSI for the variable
        psi = calculate_psi_bucket(expected_counts, actual_counts)
        
        psi_values[col] = psi
    
    return psi_values

# Calculate PSI for each variable between df_1 and df_2
psi_values = calculate_psi(df_train[x_col], df_test[x_col])

# Visualize PSI values
psi_df = pd.DataFrame.from_dict(psi_values, orient='index', columns=['PSI'])
psi_df = psi_df.sort_values(by='PSI', ascending=False)

# Plot the PSI values
plt.figure(figsize=(12, 6))
sns.barplot(x=psi_df.index, y='PSI', data=psi_df, palette='viridis')
plt.xticks(rotation=90)
plt.xlabel('Variable')
plt.ylabel('Population Stability Index (PSI)')
plt.title('Population Stability Index (PSI) for Variables')
plt.tight_layout()
plt.show()
```
